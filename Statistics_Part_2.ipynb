Q1- What is hypothesis testing in statistics?
- Answer Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It involves evaluating two competing statements‚Äîcalled hypotheses‚Äîabout a population parameter.

Q2- What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Answer The null hypothesis (H‚ÇÄ) and the alternative hypothesis (H‚ÇÅ or Ha) are foundational to hypothesis testing, and they represent opposing statements about a population parameter.
Key Differences:
Feature	                Null Hypothesis (H‚ÇÄ)	Alternative Hypothesis (H‚ÇÅ)
Assumption	            No effect/difference	There is an effect/difference

Q3- What is the significance level in hypothesis testing, and why is it important?
- Answer The significance level, denoted by Œ± (alpha), is a threshold used in hypothesis testing to determine whether the observed data is statistically significant‚Äîthat is, whether you should reject the null hypothesis (H‚ÇÄ).

Why Is It Important?
Controls the Risk of False Positives
By setting Œ±, you control how willing you are to make a wrong decision by rejecting a true null hypothesis.

Sets the Decision Boundary
If the p-value ‚â§ Œ±, you reject H‚ÇÄ (evidence is statistically significant).
If the p-value > Œ±, you fail to reject H‚ÇÄ (not enough evidence).

Determines the Strength of Evidence Required
A smaller Œ± (e.g., 0.01) means you need stronger evidence to reject H‚ÇÄ, while a larger Œ± (e.g., 0.10) means you're more lenient.

Q4- What does a P-value represent in hypothesis testing?
- Answer A p-value in hypothesis testing represents the probability of obtaining results as extreme as (or more extreme than) the observed data, assuming the null hypothesis (H‚ÇÄ) is true.

Q6- What are Type 1 and Type 2 errors in hypothesis testing?
- Answer In hypothesis testing, Type I and Type II errors represent two kinds of mistakes you can make when drawing conclusions from data.
Type I Error (False Positive)
Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

Think: You concluded there's an effect when there isn't one.

Probability: Denoted by Œ± (alpha) ‚Äî the significance level, e.g., 0.05.

Consequence: You act on a result that was just due to random chance.

Example:
You test a drug that doesn‚Äôt actually work but conclude it does ‚Äî approving an ineffective treatment.

üîπ Type II Error (False Negative)
Definition: Failing to reject the null hypothesis when it is actually false.

Think: You missed a real effect.

Probability: Denoted by Œ≤ (beta).

Power of a test: Equal to 1 ‚àí Œ≤ (the chance you correctly reject H‚ÇÄ when it's false).

Example:
A new teaching method is genuinely better, but your test doesn't detect the difference, so you incorrectly conclude it has no effect.

```python
## Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
from scipy import stats
import numpy as np

# Sample data
sample = np.array([52, 49, 50, 48, 51, 53, 47, 50, 49, 52])
population_mean = 50
population_std = 2  # known standard deviation

# Z-test
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed

print(f"Z-score: {z_score:.3f}, P-value: {p_value:.3f}")
```

```python
##  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
def visualize_errors(mu_0=0, mu_a=1, sigma=1, alpha=0.05):
    import seaborn as sns

    x = np.linspace(-4, 4, 1000)
    z_crit = stats.norm.ppf(1 - alpha)

    y0 = stats.norm.pdf(x, loc=mu_0, scale=sigma)
    y1 = stats.norm.pdf(x, loc=mu_a, scale=sigma)

    plt.figure(figsize=(10,5))
    plt.plot(x, y0, label='H‚ÇÄ: Œº=0')
    plt.plot(x, y1, label='H‚ÇÅ: Œº=1')
    plt.fill_between(x, y0, where=(x >= z_crit), color='red', alpha=0.3, label='Type I Error (Œ±)')
    plt.fill_between(x, y1, where=(x < z_crit), color='blue', alpha=0.3, label='Type II Error (Œ≤)')
    plt.axvline(z_crit, color='black', linestyle='--')
    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.show()

visualize_errors()
```

```python
## Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array): Sample data points
        confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
        (float, float): Lower and upper bounds of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * sem
    return mean - margin, mean + margin
```

```python
## Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
        float: Margin of error
    """
    n = len(data)
    sem = stats.sem(data)  # standard error of the mean
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_score * sem
    return margin

# Example usage:
sample_data = [52, 55, 49, 50, 51, 54, 53, 48, 50, 52]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level*100)}% confidence: ¬±{moe:.2f}")
```

```python
## Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, false_positive, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): The prior probability (P(H))
        likelihood (float): The likelihood (P(E|H))
        false_positive (float): The false positive rate (P(E|¬¨H))
        evidence (float): The probability of evidence (P(E))

    Returns:
        float: The posterior probability (P(H|E))
    """
    # P(E) is the total probability of the evidence, considering both possibilities
    # P(E) = P(E|H) * P(H) + P(E|¬¨H) * P(¬¨H)
    not_prior = 1 - prior  # P(¬¨H) = 1 - P(H)

    evidence_probability = (likelihood * prior) + (false_positive * not_prior)

    # Bayes' Theorem: P(H|E) = P(E|H) * P(H) / P(E)
    posterior = (likelihood * prior) / evidence_probability
    return posterior

# Given data
prior_probability = 0.01          # Prior probability P(H) (Person is sick)
likelihood = 0.99                # Likelihood P(E|H) (Test is positive if sick)
false_positive = 0.05            # False positive rate P(E|¬¨H) (Test is positive if not sick)
evidence = None                  # This will be calculated inside the function

# Compute posterior probability using Bayes' Theorem
posterior_probability = bayesian_inference(prior_probability, likelihood, false_positive, evidence)
print(f"The posterior probability that the person is sick given they tested positive is: {posterior_probability*100:.2f}%")
```

```python
## Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a sample dataset: Gender and Product Preference
data = {
    'Product A': [30, 10],  # [Male, Female]
    'Product B': [10, 30],  # [Male, Female]
    'Product C': [5, 15]    # [Male, Female]
}

# Convert into a DataFrame (Contingency Table)
df = pd.DataFrame(data, index=['Male', 'Female'])

# Perform the Chi-square test for independence
chi2, p, dof, expected = chi2_contingency(df)

# Output the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:\n", expected)

# Interpret the result
alpha = 0.05
if p < alpha:
    print(f"Reject the null hypothesis (p-value = {p:.4f} < {alpha})")
    print("Conclusion: There is a significant relationship between gender and product preference.")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p:.4f} > {alpha})")
    print("Conclusion: There is no significant relationship between gender and product preference.")
```

```python
## Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
        observed_data (ndarray): 2D array representing the observed contingency table.

    Returns:
        ndarray: 2D array representing the expected frequencies.
    """
    # Calculate the row totals, column totals, and grand total
    row_totals = np.sum(observed_data, axis=1)
    col_totals = np.sum(observed_data, axis=0)
    grand_total = np.sum(observed_data)

    # Calculate the expected frequencies
    expected_data = np.outer(row_totals, col_totals) / grand_total
    return expected_data

# Example observed data: A 2x3 contingency table (for example, Gender vs Product Preference)
observed_data = np.array([
    [30, 10, 5],  # Male responses
    [10, 30, 15]  # Female responses
])

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the expected frequencies
print("Observed Data:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
##  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Example observed data (observed frequencies for each category)
observed_data = np.array([50, 60, 70, 80, 40])  # Observed frequencies for 5 categories

# Expected frequencies (assuming a uniform distribution)
expected_data = np.array([60, 60, 60, 60, 60])  # Expected frequencies for a uniform distribution

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed_data, expected_data)

# Output the results
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"Reject the null hypothesis (p-value = {p_value:.4f} < {alpha})")
    print("Conclusion: The observed data does not follow the expected distribution.")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p_value:.4f} > {alpha})")
    print("Conclusion: The observed data follows the expected distribution.")
```

```python
##  Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Function to simulate and visualize the Chi-square distribution
def visualize_chi_square(df_values, size=10000):
    """
    Simulate and visualize the Chi-square distribution for various degrees of freedom.

    Parameters:
        df_values (list): A list of degrees of freedom to visualize.
        size (int): Number of random samples to generate for simulation.
    """
    plt.figure(figsize=(10, 6))

    # Generate and plot the Chi-square distributions for different degrees of freedom
    for df in df_values:
        # Simulate random data from the Chi-square distribution
        data = np.random.chisquare(df, size)
        # Plot the histogram for the simulated data
        plt.hist(data, bins=30, alpha=0.5, label=f'df={df}', density=True)

    # Add theoretical Chi-square distribution curves for comparison
    x = np.linspace(0, max(df_values) * 2, 200)
    for df in df_values:
        plt.plot(x, chi2.pdf(x, df), label=f'Theoretical (df={df})', linewidth=2)

    plt.title('Chi-square Distribution for Various Degrees of Freedom')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

# List of degrees of freedom to visualize
df_values = [1, 2, 5, 10, 20]

# Call the function to visualize the Chi-square distribution
visualize_chi_square(df_values)
```

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two random samples.

    Parameters:
        sample1 (array-like): The first sample data.
        sample2 (array-like): The second sample data.
        alpha (float): Significance level for the test.

    Returns:
        f_stat (float): The computed F-statistic.
        p_value (float): The p-value for the F-test.
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate degrees of freedom for both samples
    n1, n2 = len(sample1), len(sample2)
    df1, df2 = n1 - 1, n2 - 1

    # Compute the F-statistic (variance ratio)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Calculate the p-value using the F-distribution
    p_value = 1 - f.cdf(f_stat, df1, df2)

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis (p-value = {p_value:.4f} < {alpha})")
        print("Conclusion: The variances are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis (p-value = {p_value:.4f} > {alpha})")
        print("Conclusion: The variances are not significantly different.")

    return f_stat, p_value

# Example random samples
sample1 = np.random.normal(loc=10, scale=5, size=100)  # Sample 1 with mean=10, std=5
sample2 = np.random.normal(loc=10, scale=7, size=100)  # Sample 2 with mean=10, std=7

# Perform the F-test
f_statistic, p_value = f_test(sample1, sample2, alpha=0.05)
```

```python
## Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Example data: Test scores based on Teaching Method and Study Time
# Factor 1: Teaching Method (3 levels)
# Factor 2: Study Time (2 levels)

np.random.seed(42)

# Generating sample data
teaching_method = ['Method A', 'Method B', 'Method C']
study_time = ['Short', 'Long']

# Creating a dataframe for test scores with 3 teaching methods and 2 study time levels
data = []

for method in teaching_method:
    for time in study_time:
        # Simulating test scores (with some noise)
        scores = np.random.normal(loc=75 + 5*teaching_method.index(method) + 2*(study_time.index(time)), scale=10, size=10)
        for score in scores:
            data.append([method, time, score])

df = pd.DataFrame(data, columns=['Teaching_Method', 'Study_Time', 'Test_Score'])

# Performing Two-Way ANOVA using the OLS (Ordinary Least Squares) model
model = ols('Test_Score ~ C(Teaching_Method) + C(Study_Time) + C(Teaching_Method):C(Study_Time)', data=df).fit()

# Getting ANOVA table
anova_results = anova_lm(model)
print(anova_results)

# Visualization of the results
plt.figure(figsize=(10, 6))

# Interaction plot to visualize interaction between Teaching Method and Study Time
sns.boxplot(x='Teaching_Method', y='Test_Score', hue='Study_Time', data=df)
plt.title('Interaction between Teaching Method and Study Time on Test Scores')
plt.xlabel('Teaching Method')
plt.ylabel('Test Scores')
plt.show()
```

```python
## Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Function to visualize the F-distribution for different degrees of freedom
def visualize_f_distribution(df1_values, df2_values, x_range=(0, 5)):
    """
    Visualizes the F-distribution for different degrees of freedom (df1, df2).

    Parameters:
    - df1_values: List of degrees of freedom for the numerator.
    - df2_values: List of degrees of freedom for the denominator.
    - x_range: Range of x values for plotting the F-distribution (default: 0 to 5).
    """
    x = np.linspace(x_range[0], x_range[1], 1000)

    plt.figure(figsize=(10, 6))

    for df1 in df1_values:
        for df2 in df2_values:
            y = f.pdf(x, df1, df2)  # Calculate the PDF for the F-distribution
            plt.plot(x, y, label=f'df1={df1}, df2={df2}')

    plt.title("F-Distribution for Different df1 and df2", fontsize=16)
    plt.xlabel("F-value", fontsize=14)
    plt.ylabel("Probability Density", fontsize=14)
    plt.legend(title="Degrees of Freedom")
    plt.grid(True)
    plt.show()

# Example: Visualizing the F-distribution with varying df1 and df2
df1_values = [2, 5, 10]  # Degrees of freedom for the numerator
df2_values = [2, 10, 30]  # Degrees of freedom for the denominator

visualize_f_distribution(df1_values, df2_values)
```

```python
## Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Simulating data: Test scores for three teaching methods
np.random.seed(42)

# Teaching methods
methods = ['Method A', 'Method B', 'Method C']

# Simulated test scores for each method
scores_method_A = np.random.normal(loc=75, scale=10, size=30)  # mean=75, std=10, n=30
scores_method_B = np.random.normal(loc=80, scale=10, size=30)  # mean=80, std=10, n=30
scores_method_C = np.random.normal(loc=85, scale=10, size=30)  # mean=85, std=10, n=30

# Create a DataFrame
data = {
    'Teaching_Method': ['Method A']*30 + ['Method B']*30 + ['Method C']*30,
    'Test_Score': np.concatenate([scores_method_A, scores_method_B, scores_method_C])
}

df = pd.DataFrame(data)

# Perform One-Way ANOVA
f_statistic, p_value = f_oneway(scores_method_A, scores_method_B, scores_method_C)

# Print the ANOVA results
print("ANOVA Test Results:")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Visualize the data using boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(x='Teaching_Method', y='Test_Score', data=df)
plt.title('One-Way ANOVA: Comparison of Test Scores Across Different Teaching Methods', fontsize=16)
plt.xlabel('Teaching Method', fontsize=14)
plt.ylabel('Test Score', fontsize=14)
plt.grid(True)
plt.show()

# Interpret the results
if p_value < 0.05:
    print("We reject the null hypothesis: There is a significant difference between the teaching methods.")
else:
    print("We fail to reject the null hypothesis: There is no significant difference between the teaching methods.")
```

```python
##  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from normal distribution
np.random.seed(42)  # Set seed for reproducibility

# Parameters for the control and treatment groups
control_mean = 50
control_std = 10
treatment_mean = 55
treatment_std = 10

# Sample size for each group
n = 100

# Simulate data
control_group = np.random.normal(loc=control_mean, scale=control_std, size=n)
treatment_group = np.random.normal(loc=treatment_mean, scale=treatment_std, size=n)

# Step 2: Perform a two-sample T-test
t_statistic, p_value = stats.ttest_ind(control_group, treatment_group)

# Step 3: Interpret the results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("We reject the null hypothesis: The means of the two groups are significantly different.")
else:
    print("We fail to reject the null hypothesis: The means of the two groups are not significantly different.")

# Step 4: Visualize the data with boxplots
plt.figure(figsize=(10, 6))
plt.boxplot([control_group, treatment_group], labels=["Control", "Treatment"])
plt.title('Comparison of Control and Treatment Groups')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
##  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Step 1: Simulate sample data
np.random.seed(42)

# Parameters
population_variance = 25  # Hypothesized population variance
sample_size = 30  # Sample size
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate sample variance (s^2)
sample_variance = np.var(sample_data, ddof=1)

# Step 3: Perform the Chi-square test
# Test statistic formula: (n - 1) * s^2 / sigma_0^2
test_statistic = (sample_size - 1) * sample_variance / population_variance

# Step 4: Determine the critical values for a two-tailed test (alpha = 0.05)
alpha = 0.05
df = sample_size - 1
critical_value_lower = chi2.ppf(alpha / 2, df)
critical_value_upper = chi2.ppf(1 - alpha / 2, df)

# Step 5: Interpret the results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Test Statistic (Chi-squared): {test_statistic:.4f}")
print(f"Critical values: Lower = {critical_value_lower:.4f}, Upper = {critical_value_upper:.4f}")

# Interpretation
if test_statistic < critical_value_lower or test_statistic > critical_value_upper:
    print("We reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("We fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")

# Step 6: Visualize the Chi-square distribution and the test statistic
x = np.linspace(0, 50, 500)
y = chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square Distribution (df={df})')
plt.axvline(test_statistic, color='r', linestyle='--', label=f'Test Statistic = {test_statistic:.4f}')
plt.axvline(critical_value_lower, color='g', linestyle='--', label=f'Critical Lower Value = {critical_value_lower:.4f}')
plt.axvline(critical_value_upper, color='g', linestyle='--', label=f'Critical Upper Value = {critical_value_upper:.4f}')

plt.title('Chi-square Test for Population Variance', fontsize=16)
plt.xlabel('Chi-square Value', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
##  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
from scipy.stats import norm

# Step 1: Simulate data for two groups
# Group 1: 1000 trials with 550 successes
n1 = 1000
x1 = 550
p1 = x1 / n1  # Proportion of successes for Group 1

# Group 2: 1200 trials with 600 successes
n2 = 1200
x2 = 600
p2 = x2 / n2  # Proportion of successes for Group 2

# Step 2: Calculate the pooled proportion (P)
P = (x1 + x2) / (n1 + n2)

# Step 3: Calculate the Z-test statistic
z = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Step 4: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 5: Interpret the results
print(f"Proportion of Successes in Group 1: {p1:.4f}")
print(f"Proportion of Successes in Group 2: {p2:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("We reject the null hypothesis: There is a significant difference between the two proportions.")
else:
    print("We fail to reject the null hypothesis: There is no significant difference between the two proportions.")
```

```python
## Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets with different variances
np.random.seed(42)

# Group 1: 100 samples with variance 25 (std deviation = 5)
n1 = 100
var1 = 25
data1 = np.random.normal(loc=0, scale=np.sqrt(var1), size=n1)

# Group 2: 100 samples with variance 50 (std deviation = 7.07)
n2 = 100
var2 = 50
data2 = np.random.normal(loc=0, scale=np.sqrt(var2), size=n2)

# Step 2: Calculate sample variances
s1_squared = np.var(data1, ddof=1)
s2_squared = np.var(data2, ddof=1)

# Step 3: Calculate the F-test statistic (larger variance in the numerator)
F_statistic = s1_squared / s2_squared if s1_squared > s2_squared else s2_squared / s1_squared

# Step 4: Degrees of freedom
df1 = n1 - 1  # Degrees of freedom for group 1
df2 = n2 - 1  # Degrees of freedom for group 2

# Step 5: Calculate the p-value using the F-distribution
p_value = 2 * min(f.cdf(F_statistic, df1, df2), 1 - f.cdf(F_statistic, df1, df2))

# Step 6: Interpret the results
print(f"Sample Variance Group 1: {s1_squared:.4f}")
print(f"Sample Variance Group 2: {s2_squared:.4f}")
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("We reject the null hypothesis: The variances of the two groups are significantly different.")
else:
    print("We fail to reject the null hypothesis: The variances of the two groups are not significantly different.")

# Step 7: Visualize the F-distribution and test statistic
x = np.linspace(0, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
plt.axvline(F_statistic, color='r', linestyle='--', label=f'F-statistic = {F_statistic:.4f}')
plt.title('F-test for Comparing Variances', fontsize=16)
plt.xlabel('F-statistic', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
